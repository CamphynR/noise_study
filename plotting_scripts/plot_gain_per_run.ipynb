```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def convert_to_db(gain):
    db = 20*np.log10(gain)
    return db
```

```python
station_id = 11
season = 2023
calibration_per_run_dir = f"/user/rcamphyn/noise_study/absolute_amplitude_results/results_per_run/season{season}/station{station_id}"
```

```python
calibration_files = glob.glob(calibration_per_run_dir + "/*")
```

```python
dfs = [pd.read_csv(file) for file in calibration_files]
```

```python
df = pd.concat(dfs)
```

```python
channel_id = 0
```

```python
df_ch = df.loc[channel_id]
```

```python
gains = df_ch["gain"].values
```

```python
plt.style.use("retro")
fig, ax = plt.subplots()
ax.hist(convert_to_db(gains));
ax.set_xlabel("Gain / dB");
ax.set_ylabel("Counts");
ax.set_title(f"Season {season} station {station_id} channel {channel_id}");
```
