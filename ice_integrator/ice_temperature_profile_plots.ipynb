```python
import numpy as np
import matplotlib.pyplot as plt
```

# Temperature profile and attenuation length

```python
from temperature_profiles import temperature_profile_grip
grip = temperature_profile_grip()
```

```python
plt.plot(grip.profile(np.linspace(0,-3000,1000)), np.linspace(0,-3000,1000), label="GRIP profile")
plt.xlabel(r"temperature [$^{\circ}$C]")
plt.ylabel("z [m]")
plt.legend()
plt.savefig("grip_temperature_vs_depth.pdf")
```

```python
plt.plot(np.linspace(-5, -35, 10), grip.attenuation_from_temperature(np.linspace(-5, -35, 10)), label="GRIP profile")
plt.semilogy()
plt.xlim(-35, -5)
plt.ylim(1e2, 1e4)
plt.xlabel(r"temperature [$^{\circ}$C]")
plt.ylabel("field attenuation length [m]")
plt.legend()
plt.savefig("grip_attenuation_vs_temperature.pdf")
```

```python
depths = np.linspace(0,-3000,301)
atts = grip.attenuation_from_depth(depths)
plt.plot(atts, depths)
plt.axvline(np.mean(atts[depths>-2000]), linestyle="--", color="black", label=fr"<L(att)> above -2000m ({int(np.mean(atts[depths>-2000]))} m)")
plt.ylabel(r"z [m]")
plt.xlabel("field attenuation length [m]")
plt.legend()
plt.ylim(-3000, 0)
plt.xlim(0,1300)
plt.savefig("grip_attenuation_vs_depth.pdf")
```
